```python
import networkx as nx

G = nx.karate_club_graph()
G = nx.convert_node_labels_to_integers(G, first_label=1)
degCent = nx.degree_centrality(G)
degCent[34]
```

```python
closeCent = nx.closeness_centrality(G)
closeCent[34]
```

```python
closeCent[32]
```

```python
G1 = nx.Graph()
G1.add_edges_from([('A','B'),('A','C'),('B','D'),('C','D'),('C','E'),('D','G'),('D','E'),('E','G'),('G','F')])
nx.draw_networkx(G1)
```

```python
nx.degree_centrality(G1)
```

```python
nx.closeness_centrality(G1,'G')
```

```python
nx.betweenness_centrality(G1, k =None, normalized=True)
```

```python
nx.edge_betweenness_centrality(G1, k=None, normalized=False)
```

```python
G2 = nx.DiGraph()
G2.add_edges_from([('B','A'),
                  ('A','B'),
                  ('A','C'),
                  ('C','D'),
                  ('D','C')])
nx.draw_networkx(G2)
```

```python
for x in [0.8, 0.9, 0.95, 0.5]:
    print(nx.pagerank(G2, alpha = x))
```

```python
G3 = nx.DiGraph()
G3.add_edges_from([('A','B'),
                  ('A','C'),
                  ('B','C'),
                  ('C','A'),
                  ('D','C')])
nx.draw_networkx(G3)
```

```python
nx.pagerank(G3, alpha=1)
```
